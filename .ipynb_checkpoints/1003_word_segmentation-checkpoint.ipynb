## train syllable model (BI)

```python
# load data and build dict

datasetname = 'sy_1'

with open('dataset/' + datasetname + '_train.json') as file: train_data = json.load(file)
with open('dataset/' + datasetname + '_validate.json') as file: validate_data = json.load(file)
with open('dataset/' + datasetname + '_new_test.json') as file: test100_data = json.load(file)
with open ('dataset/' + datasetname + '_dict.json') as file: _to_ix = json.load(file)
print('size of data:', len(train_data), len(validate_data))
x1_to_ix, tag_to_ix = _to_ix
```

```python
model = BiLSTM_CRF(len(x1_to_ix), label_size=len(tag_to_ix), x1emb_dim=512, hidden_dim=1024).cuda()
```

```python
best_score = train(model, _to_ix, 3, bs=160, learn_rate=.001, 
                  train_data=train_data[:1000], validate_data=train_data[:1000], best_score=0)
```

```python
# evaluate
result, pred_text, answer_text = evaluate(model, _to_ix, test100_data, bs=160)
```

```python
# train

name_to_save = 'sy_1'

# train 10 epoch at lr 0.001
best_score = train(5, 
                  bs=160, 
                  learn_rate=.001, 
                  train_data=train_data, 
                  validate_data=validate_data, 
                  best_score=0, 
                  name_to_save=name_to_save)

# load best model
model.load_state_dict(torch.load('model/' + name_to_save + '.pth'))
with open('model/' + name_to_save + '-dicts.json', 'r') as file:
    (x1_to_ix, tag_to_ix) = json.load(file)
model.train()

# train 5 epoch at lr=0.0005
best_score = train(5, 
                  bs=160, 
                  learn_rate=.0005, 
                  train_data=train_data, 
                  validate_data=validate_data, 
                  best_score=best_score, 
                  name_to_save=name_to_save)
```

```python
with open ('pred_text_BI.txt', 'w') as file:
    file.write(pred_text)
with open ('answer_text.txt', 'w') as file:
    file.write(answer_text)
```

## train syllable model (Scheme B)

```python
# load data and build dict

datasetname = 'sy_1234'

with open('dataset/' + datasetname + '_train.json') as file: train_data = json.load(file)
with open('dataset/' + datasetname + '_validate.json') as file: validate_data = json.load(file)
with open('dataset/' + datasetname + '_new_test.json') as file: test100_data = json.load(file)
print('size of data:', len(train_data), len(validate_data))
    
x1_to_ix, tag_to_ix = get_vocabs(train_data)
ix_to_tag = {ix: tag for tag, ix in tag_to_ix.items()} # get a reversed dict mapping int to str
```

```python
# reduce vocab to frequency > 1

vocab = []
for data in train_data:
    for sy in data[0]:
        vocab.append(sy)
sc = Counter(vocab)

newx1_to_ix = {Const.UNK_TOKEN: Const.UNK_ID, Const.PAD_TOKEN: Const.PAD_ID, Const.BOS_TOKEN: Const.BOS_ID, Const.EOS_TOKEN: Const.EOS_ID}
for k,v in x1_to_ix.items():
    if sc[k] > 1:
        if k not in newx1_to_ix: newx1_to_ix[k] = len(newx1_to_ix)

x1_to_ix = newx1_to_ix
```

### with dropout

```python
model = BiLSTM_CRF(len(x1_to_ix), label_size=len(tag_to_ix), x1emb_dim=512, hidden_dim=1024).cuda()
```

```python
# train

name_to_save = 'sy_1234'

# train 10 epoch at lr 0.001
best_score = train(5, 
                  bs=160, 
                  learn_rate=.001, 
                  train_data=train_data, 
                  validate_data=validate_data, 
                  best_score=0, 
                  name_to_save=name_to_save)

# load best model
model.load_state_dict(torch.load('model/' + name_to_save + '.pth'))
with open('model/' + name_to_save + '-dicts.json', 'r') as file:
    (x1_to_ix, tag_to_ix) = json.load(file)
model.train()

# train 5 epoch at lr=0.0005
best_score = train(5, 
                  bs=160, 
                  learn_rate=.0005, 
                  train_data=train_data, 
                  validate_data=validate_data, 
                  best_score=best_score, 
                  name_to_save=name_to_save)
```

```python
# train 5 epoch at lr=0.0005
best_score = train(5, 
                  bs=160, 
                  learn_rate=.0002, 
                  train_data=train_data, 
                  validate_data=validate_data, 
                  best_score=95.90, 
                  name_to_save=name_to_save)
```

```python
# evaluate
result, pred_text, answer_text = evaluate(bs=160)
```

```python
with open ('pred_text_Scheme_B.txt', 'w') as file:
    file.write(pred_text)
with open ('answer_text.txt', 'w') as file:
    file.write(answer_text)
```

## dropout results

```python
ได้ผลออกมาแล้วนะคะ

มันออกมาแปลกๆ คือไม่มี embedding dropout มันได้ f1 ดีกว่าอีก
แต่จินคิดว่า epoch ไม่พอมากกว่า 
ถ้า train ต่อไปอีกอาจจะได้พอๆกัน

แล้วตัวที่ไม่มี dropout ก็แย่ไปเลย ภายใน 10 epochs เท่ากัน
แต่ถ้า train ต่อไปนานๆอาจจะดีกว่านี้
dropout คงช่วยให้ได้ผลที่ดีเร็วขึ้นด้วย
```

```python
parameters
embedding dimension         =  512
hidden dimension            = 1024
embedding dropout           = 0.1
lstm output dropout         = 0.2
lr (first 5 epochs)         = 0.001  
lr (last  5 epochs)         = 0.0005 

f1WL
with both dropouts          = 0.9582
without embedding dropout   = 0.9592
without lstm output dropout = 0.9508
without both dropouts       = 0.8945
```
